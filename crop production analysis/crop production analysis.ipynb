### season wise yearly production

```python
num=1

for i in df2.Season.unique():
    plt.figure(figsize=(25,25))
    plt.subplot(6,2,num)
    sns.barplot(data=df2[df2["Season"]==i],x="Crop_Year",y="Production")
    plt.xticks(rotation=90)
    plt.title(i)
    plt.show()
    num+=1
    
```

During kharif season the in 2015 there was maximum production 

winter  crops  has very low production of 2004 to 2015 and similarly trend in autumn

whole year crop does not vast difference
